```python
from empiricaldist import Cdf
```

```python
# Creamos la CDF de una variable

cdf_var_name = Cdf.from_seq(
    df.flipper_length_mm
);

cdf_var_name
```

```python
# Ploteamos la CDF de la variable

cdf_var_name.plot();
```

```python
# ¿Cuál es la probabilidad (p) de que la long. de la aleta sea de 200mm o menos (q)?

q = 200
p = cdf_var_name.forward(q)

print(q, p)
```

```python
# La misma pregunta pero con DataViz

cdf_var_name.plot()

q = 200
p = cdf_var_name.forward(q)

# plotear línea vertical

plt.vlines(
    x = q,
    ymin = 0,
    ymax = p,
    color = 'black',
    linestyle = 'dashed'
)

# plotear línea horizontal

plt.hlines(
    y = p,
    xmin = pmf_var_name.qs[0],
    xmax = q,
    color = 'black',
    linestyle = 'dashed'
)

plt.plot(q, p, 'ro');
```

```python
# Plotear el Rango Intercuartil (IQR)

cdf_var_name.step()
p_1 = 8.25 # Specify probability
P_2 = 8.75
ps = (0.25, 0.75) # IQR
qs = cdf_var_name.inverse(ps)

plt.vlines (
    x = qs,
    ymin = 0,
    ymax = ps,
    color = 'black',
    linestyle = 'dashed'
)

plt.hlines(
   y = ps,
   xmin = pmf_var_name.qs[0],
   xmax = qs,
   color = 'black',
   linestyle = 'dashed'
)

plt.scatter(
   x = qs,
   y = ps,
   color = 'red',
   zorder = 2
);
```

## Comparar la distribución de nuestros datos vs una dist. norm. aleatoria

```python
import numpy as np
import scipy
from scipy import stats
from scipy.stats import norm
```

```python
# Comparando Funciones de Distribución Acumulada (CDF)

stats = df.body_mass_g.describe()

xs = np.linspace(stats['min'], stats['max'])
ys = scipy.stats.norm(stats['mean'], stats['std']).cdf(xs)

plt.plot(
    xs,
    ys,
    color = 'black',
    linestyle = 'dashed'
)

Cdf.from_seq(
    df.body_mass_g
).plot();
```

```python
# Comparando Funciones de Densidad de Probabilidad (KDE)

xs = np.linspace(stats['min'], stats['max'])
ys = scipy.stats.norm(stats['mean'], stats['std']).pdf(xs) # Aquí se cambio la función

plt.plot(
    xs,
    ys,
    color = 'black',
    linestyle = 'dashed'
)

# La diferencia es que empiricaldist no puede y Seaborn si
sns.kdeplot(
    data = df,
    x = 'body_mass_g'
);
```

Se puede apreciar que nuestros datos están sesgados hacia la izquierda

## Teorema del Limite Central

```python
df.sex.value_counts(normalize = True)
```

```python
sex_num = df.sex.replace(['male', 'female'], [1, 0])
sex_num
```

De una muestra de 1000 tomamos una muestra de 35

```python
import random

# Set variables
number_samples = 1000
sample_size = 35
samples_df = pd.DataFrame()

# Set random seed
random.seed(42)

# Create samples and add to dataframe
for i in range(1, number_samples+1):
    sex_numeric_sample = sex_num.sample(sample_size, replace=True).to_numpy()
    sample_name = f"Sample {i}"
    samples_df = pd.concat([samples_df, pd.DataFrame(sex_numeric_sample, columns=[sample_name])],axis=1)

# Calculate estimated percentage of male penguins in population
male_population_mean = samples_df.mean().mean()
print(f"Estimated percentage of male penguins in population is: {male_population_mean * 100:.2f}%")
```

Verificamos visualmente el resultado numerico

```python
sample_means_binomial = pd.DataFrame(samples_df.mean(), columns=['sample_mean'])
sns.kdeplot(data=sample_means_binomial)
plt.axvline(x=sex_num.mean(), color='red', linestyle='dashed');
```

Otra manera de demostrarlo

```python
sample_size_experiment = pd.DataFrame(
    [[i, samples_df.iloc[:, 0:i].mean().mean().mean()] for i in range(1, number_samples + 1)],
    columns=['sample_size', 'estimated_mean']
)
```

```python
sns.scatterplot(
    data=sample_size_experiment,
    x='sample_size',
    y='estimated_mean'
)

plt.axhline(
    y=sex_num.mean(),
    color='red',
    linestyle='dashed'
)

plt.ylim([sex_num.mean() - 0.20, sex_num.mean() + 0.20]);
```

## Análisis de Regresión Simple

```python
# Visualizamos 2 conjuntos de datos aleatorios

np.random.seed(42)
x1 = np.linspace(0,100,100)
y1 = 0.1* x1 +3 + np.random.uniform(-2,2,size=x1.size)

sns.scatterplot(x=x1, y=y1)


x2 = np.linspace(0,100,100)
y2 = 0.5* x1 +1 + np.random.uniform(0,60,size=x2.size)

sns.scatterplot(x=x2, y=y2)

plt.legend(["1","2"])

print(np.corrcoef(x1,y1))
print(np.corrcoef(x2,y2))
```

```python
# Calculamos 2 modelos de regresión lineal

res_1 = scipy.stats.linregress(x = x1, y = y1)
res_2 = scipy.stats.linregress(x = x2, y = y2)

print(res_1, res_2, sep = '\n')
```

```python
# Ploteamos nuestros 2 modelos de regresión lineal

sns.scatterplot(x=x1, y=y1)

fx1 = np.array([x1.min(), x1.max()])
fy1 = res_1.intercept + res_1.slope * fx1

plt.plot(fx1, fy1)


sns.scatterplot(x=x2, y=y2)

fx2 = np.array([x2.min(), x2.max()])
fy2 = res_2.intercept + res_2.slope * fx2

plt.plot(fx2, fy2)

plt.legend(["1", "1","2","2"]);
```

## Análisis Multivariado

```python
import statsmodels.formula.api as smf

model_1 = (
    smf.ols(
        formula = 'body_mass_g ~ bill_length_mm',
        data = df
    )
    .fit()
)

model_1.summary()
```

```python
model_2 = (
    smf.ols(
        formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm',
        data = df
    )
    .fit()
)

model_2.summary()
```

```python
model_3 = (
    smf.ols(
        formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm + flipper_length_mm',
        data = df
    )
    .fit()
)

model_3.summary()
```

```python
model_4 = (
    smf.ols(
        formula = 'body_mass_g ~ bill_length_mm + bill_depth_mm + flipper_length_mm + C(sex)',
        data = df
    )
    .fit()
)

model_4.summary()
```

```python
model_5 = (
    smf.ols(
        formula = 'body_mass_g ~ flipper_length_mm + C(sex)',
        data = df
    )
    .fit()
)

model_5.summary()
```

Creación de tabla de resultados

```python
models_result = pd.DataFrame(
    dict(
        actual_value = df.body_mass_g,
        prediction_model_1 = model_1.predict(),
        prediction_model_2 = model_2.predict(),
        prediction_model_3 = model_3.predict(),
        prediction_model_4 = model_4.predict(),
        prediction_model_5 = model_5.predict(),
        species = df.species,
        sex = df.sex
    )
)

models_result
```

## La Paradoja de Simpson

La paradoja de Simpson es un fenómeno estadístico que ocurre cuando una tendencia que se observa en varios grupos de datos se invierte o desaparece cuando los grupos se combinan. Esto puede llevar a conclusiones erróneas o engañosas si no se tiene en cuenta la variable que determina los grupos.

La paradoja de Simpson se puede evitar analizando las relaciones causales entre las variables y no solo las correlaciones observadas.

```python
import seaborn as sns

sns.pairplot(data = df,
             hue='species',
             palette = penguin_color
);
```

```python
sns.lmplot(
    data = df,
    x = 'bill_length_mm',
    y = 'bill_depth_mm',
    hue = 'species',
    height = 10,
    palette = penguin_color
);
```
